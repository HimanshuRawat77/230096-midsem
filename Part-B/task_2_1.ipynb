```python
import os
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_informative=10,
    n_redundant=2,
    random_state=42
)

np.random.seed(42)
mask = np.random.rand(1000, 20) < 0.85
X[mask] = 0.0


dataset = pd.DataFrame(X, columns=[f"feature_{i+1}" for i in range(20)])
dataset["target"] = y


file_path = "data/sparse_toy_dataset.csv"
dataset.to_csv(file_path, index=False)

print(f"Dataset successfully generated and saved to {file_path}")
print(f"Dataset shape: {dataset.shape}")
print(f"Percentage of zeros in features: {(dataset.drop('target', axis=1) == 0).mean().mean() * 100:.2f}%")
```
